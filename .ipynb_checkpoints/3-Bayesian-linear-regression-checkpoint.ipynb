```python
# Import libraries

import pickle
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, zscore
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
import pymc3 as pm
import arviz as az
from model_utils import preprocess_housing
from sklearn.preprocessing import StandardScaler
```

```python
# Load California housing dataset

california_housing = fetch_california_housing(as_frame=True)
x_data = california_housing.data
print(x_data.shape)
y_data = california_housing.target
print(y_data.shape)
```

```python
print("Means:")
print(x_data.mean())
```

```python
# Preprocess datapoints

x, y = preprocess_housing(x_data,y_data)
print(x.shape)
print(y.shape)
X_train, X_test, y_train, y_test = train_test_split(x, y)
```

```python
# Plot how longitude and latitude affect median house price

latitudes = x['Latitude']
longitudes = x['Longitude']
plt.scatter(latitudes, longitudes, c=y*100000, alpha=0.7)
plt.colorbar()
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.show()
```

```python
import folium
from folium.plugins import HeatMap
map_hooray = folium.Map(location=[36.77,-119.41], zoom_start = 6) 

data = [[row['Latitude'],row['Longitude']] for index, row in x_data.iterrows()]
HeatMap(data, radius=13).add_to(map_hooray)
map_hooray
```

```python
# Run PyMC algorithm to get posterior distributions for each feature

num_samples = X_train.shape[0]
model = pm.Model()

with model:
    # Define priors
    intercept = pm.Normal('intercept', mu=0, sigma=20)
    MedInc = pm.Normal('MedInc', mu=0, sigma=20)
    HouseAge = pm.Normal('HouseAge', mu=0, sigma=20)
    AveRooms = pm.Normal('AveRooms', mu=0, sigma=20)
    AveBedrms = pm.Normal('AveBedrms', mu=0, sigma=20)
    Population = pm.Normal('Population', mu=0, sigma=20)
    AveOccup = pm.Normal('AveOccup', mu=0, sigma=20)
    Latitude = pm.Normal('Latitude', mu=0, sigma=20)
    Longitude = pm.Normal('Longitude', mu=0, sigma=20)
    sigma = pm.Uniform('sigma', lower=0, upper=20)

    y_est = intercept + MedInc*X_train["MedInc"] + HouseAge*X_train["HouseAge"] + AveRooms*X_train["AveRooms"] + AveBedrms*X_train["AveBedrms"] + Population*X_train["Population"] + AveOccup*X_train["AveOccup"] + Latitude*X_train["Latitude"] + Longitude*X_train["Longitude"]

    likelihood = pm.Normal('y', mu=y_est, sigma=sigma, observed=y_train)
    
    sampler = pm.NUTS()
    
    idata = pm.sample(num_samples, sampler, progressbar=True, cores=4, chains=4)
```

```python
# Save pymc results in pkl format (Used when running on lab machine)

input_d = pickle.load(open('sampler_inference.pkl', 'rb'))
pickle.dump(input_d.posterior, open("posteriors.pkl", "wb"))
pickle.dump(input_d.sample_stats, open("sample_stats.pkl", "wb"))
pickle.dump(input_d.observed_data, open("observed_data.pkl", "wb"))

# Load posteriors from pkl file
posteriors = pickle.load(open('posteriors.pkl', 'rb'))
```

```python
# Extract mean, sd, and r_hat from summary

summary = az.summary(idata, round_to=4)
summary[['mean', 'sd', 'r_hat']].T
```

```python
# Generate test set posterior samples to evaluate model

ppc_test = pm.sample_posterior_predictive(idata, model=model)

p_test_pred = ppc_test["y"].mean(axis=0)

print(az.r2_score(y_train, p_test_pred))
```

```python
# Plot each posterior distribution

pm_data = az.from_pymc3(idata)
posteriors = pm_data.posterior

features = ["MedInc" ,"HouseAge", "AveRooms", "AveBedrms", "Population", "AveOccup", "Latitude", "Longitude"]
posterior_names = ['w0', 'w1', 'w2', 'w3', 'w4', 'w5', 'w6', 'w7', 'w8']

_, ax = plt.subplots(2, 4, figsize=(40, 15))
az.plot_dist(posteriors['MedInc'], color="C1", label="MedInc", ax=ax[0][0])
az.plot_dist(posteriors['HouseAge'], color="C2", label="HouseAge", ax=ax[0][1])
az.plot_dist(posteriors['AveRooms'], color="C3", label="AveRooms", ax=ax[0][2])
az.plot_dist(posteriors['AveBedrms'], color="C4", label="AveBedrms", ax=ax[0][3])
az.plot_dist(posteriors['Population'], color="C5", label="Population", ax=ax[1][0])
az.plot_dist(posteriors['AveOccup'], color="C6", label="AveOccup", ax=ax[1][1])
az.plot_dist(posteriors['Latitude'], color="C7", label="Latitude", ax=ax[1][2])
az.plot_dist(posteriors['Longitude'], color="C8", label="Longitude", ax=ax[1][3])
plt.rcParams.update({'font.size': 25})
plt.show()
```

```python
with model:
    ppc = pm.sample_posterior_predictive(
        idata, var_names=["MedInc" ,"HouseAge", "AveRooms", "AveBedrms", "Population", "AveOccup", "Latitude", "Longitude", "y"]
    )
az.plot_ppc(az.from_pymc3(posterior_predictive=ppc, model=model));
```
